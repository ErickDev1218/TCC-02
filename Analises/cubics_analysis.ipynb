```python
import pandas as pd
```

```python
def linha_para_latex(row, idx_fpli, idx_fga, idx_fbrkga, idx_tpli, idx_opt, idx_gap, idx_D, idx_graph, idx_order):
    valores = []

    for i, v in enumerate(row):

        if i == idx_tpli:
            continue

        if "cubic" not in row[idx_graph]:
            continue

        if i == idx_fga or i == idx_fbrkga:
            if row[idx_fpli] == row[idx_fga]:
                valores.append(f"\\textbf{{{row[idx_fga]}}}") 
                continue
            if row[idx_fpli] == row[idx_fbrkga]:
                valores.append(f"\\textbf{{{row[idx_fbrkga]}}}")
                continue
        
        if i == idx_graph:
            splited = str(v).split("_")
            if len(splited) > 1:
                valores.append(f"{splited[0]}\\_{'\\_'.join(splited[1:])}")
            else:
                valores.append(str(v))
            continue

        if i == idx_D or i == idx_gap: 
            splited = str(v).split("%")
            valores.append(f"{splited[0]}\\%")
            continue

        if i == idx_opt:
            continue

        # trata F/PLI
        if i == idx_fpli:
            if row[idx_opt] == 1:
                valores.append(f"{v}*")
            else:
                valores.append(str(v))
        else:
            valores.append(str(v))

    ub = f"{row[idx_order] * 0.75:.2f}"
    valores.insert(4, ub)
    return " & ".join(valores) + r" \\"

def create_table(file_path, output_path):
    df = pd.read_csv(file_path)
    df.sort_values(by=["V(G)", "E(G)"], inplace=True)

    idx_fpli = df.columns.get_loc("F/PLI")
    idx_fga = df.columns.get_loc("F/GA")
    idx_fbrkga = df.columns.get_loc("F/BRKGA")
    idx_tpli = df.columns.get_loc("T/PLI")
    idx_opt  = df.columns.get_loc("OPT")
    idx_gap = df.columns.get_loc("GAP(%)")
    idx_D = df.columns.get_loc("D(%)")
    idx_graph = df.columns.get_loc("Graph")
    idx_order = df.columns.get_loc("V(G)")

    final = []
    for row in df.itertuples(index=False, name=None):
        final.append(linha_para_latex(row, idx_fpli, idx_fga, idx_fbrkga, idx_tpli, idx_opt, idx_gap, idx_D, idx_graph, idx_order))

    with open(output_path, "w") as f:
        f.write("\n".join(final))
```
