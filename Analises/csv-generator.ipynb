```python
def generate_csv(folder_path: Path, output_csv: str):
    all_data = []

    arquivos = list(folder_path.rglob("*.csv"))
    
    for csv_file in arquivos:
        df = pd.read_csv(csv_file)
        if df.empty:
            print(f"CSV vazio ignorado: {csv_file.name}")
            continue

        
        row = {
            "Graph": csv_file.stem,  # nome do arquivo sem .csv
            "V(G)": df["graph_order"].iloc[0],
            "E(G)": df["graph_size"].iloc[0],
            "D(%)": f"{(2 * df["graph_size"].iloc[0] / (df["graph_order"].iloc[0] * (df["graph_order"].iloc[0] - 1)) * 100):.2f}%",
            "Fitness": int(df["fitness_value"].mean()),
            "MIN" : df["fitness_value"].min(),
            "STD": f"{df["fitness_value"].std():.2f}",
            "Time(s)": f"{df["elapsed_time(seconds)"].mean():.3f}"
        }
        all_data.append(row)
        
    final_df = pd.DataFrame(all_data, columns=["Graph", "V(G)", "E(G)", "D(%)", "Fitness", "MIN", "STD", "Time(s)"])
    final_df.to_csv(output_csv, index=False)
```

```python
def generate_csv_PLI(folder_path: Path, output_csv: str):
    all_data = []

    arquivos = list(folder_path.rglob("*.csv"))
    
    for csv_file in arquivos:
        df = pd.read_csv(csv_file)
        if df.empty:
            print(f"CSV vazio ignorado: {csv_file.name}")
            continue

        
        row = {
            "Graph": csv_file.stem,  # nome do arquivo sem .csv
            "V(G)": df["graph_order"].iloc[0],
            "E(G)": df["graph_size"].iloc[0],
            "D(%)": f"{(2 * df["graph_size"].iloc[0] / (df["graph_order"].iloc[0] * (df["graph_order"].iloc[0] - 1)) * 100):.2f}%",
            "Fitness": int(df["objective_value"].mean()),
            "Time(s)": f"{min(int(df['elapsed_time(seconds)']), 900):.3f}",
            "Optimal": df[" optimal_value"].iloc[0],
        }
        all_data.append(row)
        
    final_df = pd.DataFrame(all_data, columns=["Graph", "V(G)", "E(G)", "D(%)", "Fitness", "Time(s)", "Optimal"])
    final_df.to_csv(output_csv, index=False)
```

```python
generate_csv(dimacs_GA, "dimacs_GA_results.csv")
generate_csv(HB_GA, "HB_GA_results.csv")
generate_csv(random_GA, "random_GA_results.csv")

print("BRKGA Results")

generate_csv(dimacs_BRKGA, "dimacs_BRKGA_results.csv")
generate_csv(HB_BRKGA, "HB_BRKGA_results.csv")
generate_csv(random_BRKGA, "random_BRKGA_results.csv")

print("PLI Results")
generate_csv_PLI(dimacs_PLI, "dimacs_PLI_results.csv")
generate_csv_PLI(HB_PLI, "HB_PLI_results.csv")
generate_csv_PLI(random_PLI, "random_PLI_results.csv")
```
