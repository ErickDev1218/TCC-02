```python
dimacs_GA = Path("/Users/erickdev1218/Desktop/Repositorios/TCC-02/GA-CPP/results/ga-cpp/DIMACS")
HB_GA = Path("/Users/erickdev1218/Desktop/Repositorios/TCC-02/GA-CPP/results/ga-cpp/Harwell-Boeing")
random_GA = Path("/Users/erickdev1218/Desktop/Repositorios/TCC-02/GA-CPP/results/ga-cpp/Random_graphs")


dimacs_BRKGA = Path("/Users/erickdev1218/Desktop/Repositorios/TCC-02/BRKGA/results/brkga-cpp/DIMACS")
HB_BRKGA = Path("/Users/erickdev1218/Desktop/Repositorios/TCC-02/BRKGA/results/brkga-cpp/Harwell-Boeing")
random_BRKGA = Path("/Users/erickdev1218/Desktop/Repositorios/TCC-02/BRKGA/results/brkga-cpp/Random_graphs")
```

```python
def generate_csv(folder_path: Path, output_csv: str):
    all_data = []

    arquivos = list(folder_path.rglob("*.csv"))
    print(len(arquivos))
    
    for csv_file in arquivos:
        df = pd.read_csv(csv_file)
        if df.empty:
            print(f"CSV vazio ignorado: {csv_file.name}")
            continue

        
        row = {
            "Graph": csv_file.stem,  # nome do arquivo sem .csv
            "V(G)": df["graph_order"].iloc[0],
            "E(G)": df["graph_size"].iloc[0],
            "D": 2 * df["graph_size"].iloc[0] / (df["graph_order"].iloc[0] * (df["graph_order"].iloc[0] - 1)),
            "Fitness": int(df["fitness_value"].mean()),
            "Time (seconds)": f"{df["elapsed_time(seconds)"].mean():.3f}"
        }
        all_data.append(row)
        
    final_df = pd.DataFrame(all_data, columns=["Graph", "V(G)", "E(G)", "D", "Fitness", "Time (seconds)"])
    final_df.to_csv(output_csv, index=False)
        # all_data.append(df)
```

```python
generate_csv(dimacs_GA, "dimacs_GA_results.csv")
generate_csv(HB_GA, "HB_GA_results.csv")
generate_csv(random_GA, "random_GA_results.csv")

print("BRKGA Results")

generate_csv(dimacs_BRKGA, "dimacs_BRKGA_results.csv")
generate_csv(HB_BRKGA, "HB_BRKGA_results.csv")
generate_csv(random_BRKGA, "random_BRKGA_results.csv")
```
