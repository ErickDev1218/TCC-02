```python
def merge_csv(file_list, output_file):
    GA = pd.read_csv(file_list[0])
    BRKGA = pd.read_csv(file_list[1])
    PLI = pd.read_csv(file_list[2])

    merged_df = GA.merge(BRKGA[["Graph" ,"Fitness", "MIN", "STD", "Time(s)"]], on="Graph", suffixes=('_GA', '_BRKGA'))
    pli_renamed = PLI.rename(columns=lambda c: f"{c}_PLI" if c != "Graph" else c)
    merged_df = merged_df.merge(pli_renamed[["Graph", "Fitness_PLI", "Time(s)_PLI", "Optimal_PLI"]], on="Graph")

    merged_df.rename(columns={
        "Fitness_GA" : "F/GA",
        "Fitness_BRKGA" : "F/BRKGA",
        "Fitness_PLI" : "F/PLI",
        "Optimal_PLI" : "OPT",
        "Time(s)_GA" : "T/GA",
        "Time(s)_BRKGA" : "T/BRKGA",
        "Time(s)_PLI" : "T/PLI",
        "MIN_GA": "MIN/GA",
        "MIN_BRKGA": "MIN/BRKGA",
        "STD_GA": "STD/GA",
        "STD_BRKGA": "STD/BRKGA"
    }, inplace=True)

     # ðŸ”¹ Melhor fitness entre GA e BRKGA
    merged_df["F/BEST"] = merged_df[["F/GA", "F/BRKGA"]].min(axis=1)

    # ðŸ”¹ Gap relativo em relaÃ§Ã£o ao PLI
    merged_df["GAP(%)"] = (
        (merged_df["F/BEST"] - merged_df["F/PLI"]) /
        merged_df["F/PLI"]
    ) * 100
    merged_df["GAP(%)"] = merged_df["GAP(%)"].round(2)
    
    merged_df.drop(columns=["F/BEST"], inplace=True)

    # Graph,V(G),E(G),D(%),F/GA,MIN/GA,STD/GA,T/GA,F/BRKGA,MIN/BRKGA,STD/BRKGA,T/BRKGA,F/PLI,T/PLI,OPT,GAP(%)

    merged_df = merged_df[[
        "Graph", "V(G)", "E(G)", "D(%)",
        "F/GA", 
        "F/BRKGA", 
        "F/PLI", 
        "OPT", 
        "T/GA",
        "T/BRKGA",
        "T/PLI", 
        "MIN/GA", 
        "MIN/BRKGA", 
        "STD/GA", 
        "STD/BRKGA", 
        "GAP(%)"
    ]]
    merged_df.to_csv(output_file, index=False)
```
